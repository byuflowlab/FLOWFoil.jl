# Parametric Section (PARSEC)

$$z = \sum_{i=1}^6 a_i x^{i-1/2} $$

## Parameters

There are 11 parameters in the traditional PARSEC parameterization method. They are as follows:

| Parameter | Definition |
|------|:------|
| $r_{LE}$ | Leading Edge Radius |
| $x_u$ | Chordwise position of maximum thickness of upper surface |
| $x_l$ | Chordwise position of maximum thickness of lower surface |
| $z_u$ | z-coordinate at maximum thickness of upper surface |
| $z_l$ | z-coordinate at maximum thickness of lower surface |
| $z_{xx_u}$ | Second derivative of upper surface curve at point of maximum thickness |
| $z_{xx_l}$ | Second derivative of lower surface curve at point of maximum thickness |
| $\alpha_{TE}$ | Trailing edge angle |
| $\beta_{TE}$ | Boat-tail angle |
| $z_{TE}$ | z-position of center of trailing edge |
| $\Delta z_{TE}$ | z-distance between upper and lower surface trailing edge points |

### Modified Parameters
In order to create a more intuitive parameterization, the trailing edge parameters, both position and angles are replaced to be directly the trailing edge positions of the upper and lower surface and trailing edge surface angles.

| Original Parameter | Modified Parameter | New Definition |
|------|------|:------|
| $\alpha_{TE}$ | $\theta_{TE_u}$ | Upper surface trailing edge tangent angle |
| $\beta_{TE}$ | $\theta_{TE_l}$ | Upper surface trailing edge tangent angle |
| $z_{TE}$ | $z_{TE_u}$ | z-coordinate of upper surface trailing edge |
| $\Delta z_{TE}$ | $z_{TE_l}$ | z-coordinate of lower surface trailing edge |

## Derivation of Coefficients
By inserting know elements of the airfoil geometry, and applying basic differentiation, the $a_i$ coefficients of the PARSEC model can be calculated.  

Beginning at the leading edge and moving backward, and assuming a unit airfoil we start at $x_{LE}$ demonstrating the 6 parameter-coefficient relationships for the upper surface.

Our first parameter of interest is the leading edge radius, $R_{LE}$.

The general equation for a radius of curvature (assuming a twice differentiable curve) is

$$ R = \left| \frac{\left[1+\left(\frac{dz}{dx}\right)^2\right]^{3/2}}{\frac{d^2z}{dx^2}}\right| $$

The first derivative of the PARSEC polynomial is

$$\frac{dz}{dx} = \sum_{i=1}^6 a_{i} \left(i-\frac{1}{2}\right) x^{i-3/2} $$

and the second

$$\frac{d^2z}{dx^2} = \sum_{i=1}^6 a_{i} \left(i-\frac{1}{2}\right)\left(i-\frac{3}{2}\right) x^{i-5/2} $$

plugging these into the radius of curvature equation

$$R= \left| \frac{\left( 1+ \left(\left(i-\frac{1}{2}\right)a_i x^{i-1/2}\right)^2\right)^{3/2}}{\left(i-\frac{1}{2}\right) \left(i-\frac{3}{2}\right) a_i x^{i-3/2}} \right|$$

[why does it only depend on a1?] Because the leading edge radius only depends on the coefficient $a_1$ we can replace $i$ in the equation with $1$.  We also use $x_{LE}$ as it is the leading edge only where the leading edge radius is defined.

$$R_{LE_{up}}= \left| \frac{\left( 1+ \left(\left(1-\frac{1}{2}\right)a_{1_{up}} x_{LE}^{1-1/2}\right)^2\right)^{3/2}}{\left(1-\frac{1}{2}\right) \left(1-\frac{3}{2}\right) a_{1_{up}} x_{LE}^{1-3/2}} \right|$$

Simplifying:

$$R_{LE_{up}}= \left| -\frac{x_{LE}^{3/2} \left(\frac{a_{1_{up}}^2 + 4 x_{LE}}{x_{LE}}\right)^{3/2}}{2 a_{1_{up}}} \right| $$

Canceling the $x_{LE}^{3/2}$ to avoid dividing by zero (remembering that $x_{LE} = 0$)

$$R_{LE_{up}}= \left| -\frac{\left(a_{1_{up}}^2 + 4 x_{LE}\right)^{3/2}}{2 a_{1_{up}}} \right| $$

Now inputting zero for $x_{LE}$

$$R_{LE_{up}}= \left| -\frac{\left(a_{1_{up}}^2\right)^{3/2}}{2 a_{1_{up}}} \right| $$

and simplifying down to our first relation:

$$R_{LE_{up}}= \left| -\frac{a_{1_{up}}^2}{2} \right| $$

Traveling along the upper surface of the airfoil, we find ourselves at the point of maximum thickness, defined at $(x_u,~z_u)$. (Take note that $x_u$ and $z_u$ are the parameters define above, not the x-z values of the entire upper surface curve.) Plugging that point into the PARSEC polynomial results in our second relation:

$$z_u = \sum_{i=1}^6 a_{i_{up}} x_u^{i-1/2}$$ 

At that same point, we note that the curvature there is another of our PARSEC parameters, $\frac{d^2z_{up}}{dx^2}$ or as named above: $z_{xx_u}$. Before we address the curvature relation, we notice along the way that we have an equation from the first derivative derived above, which we know equates to zero at the maximum, giving us our third relation:

$$\frac{dz_{up}}{dx} \biggr\rvert_{x=x_u} = 0 = \sum_{i=1}^6 a_{i_{up}} \left(i-\frac{1}{2}\right) x^{i-3/2} $$

where $z_{up}$ are the z-coordinates of the upper curve. Continuing on to revisit the second derivative, evaluating at the the point of maximum thickness yields our fourth relation:

$$z_{xx_u} = \sum_{i=1}^6 a_{i_{up}} \left(i-\frac{1}{2}\right)\left(i-\frac{3}{2}\right) x_u^{i-5/2}$$

As we continue along the upper surface, we arrive at the trailing edge, where our parameters define an angle ($\theta_{TE}$) and a position ($z_{TE}$). Revisiting the first derivative of our curve, we see that

$$\frac{dz_{up}}{dx} \biggr\rvert_{x=x_{TE_{up}}} = \tan\theta_{TE_{up}}$$

Taking the derivative of the PARSEC polynomial at $x_{TE} = 1$, we find our fifth relation to be:

$$ \tan\theta_{TE_{u}} = \sum_{i=1}^6 a_{i_{up}} \left(i-\frac{1}{2}\right)$$ 

Looking last at the trailing edge position, we input $x_{TE} = 1$ into the original PARSEC polynomial and arrive at our sixth and final relation:

$$z_{TE_{u}} = \sum_{i=1}^6 a_{i_{up}}$$

We represent the above relations for the upper surface in matrix form for simple computation.

$$
\begin{bmatrix}
    1 & 0 & 0 & 0 & 0 & 0 \\
    x_u^{1/2} & x_u^{3/2} & x_u^{5/2} & x_u^{7/2} & x_u^{9/2} & x_u^{11/2} \\
    \frac{1}{2}x_u^{-1/2} & \frac{3}{2}x_u^{1/2} & \frac{5}{2}x_u^{3/2} & \frac{7}{2}x_u^{5/2} & \frac{9}{2}x_u^{7/2} & \frac{11}{2}x_u^{9/2} \\
    -\frac{1}{4}x_u^{-3/2} & \frac{3}{4}x_u^{-1/2} & \frac{15}{4}x_u^{1/2} & \frac{35}{4}x_u^{3/2} & \frac{63}{4}x_u^{5/2} & \frac{99}{4}x_u^{7/2} \\
    1/2 & 3/2 & 5/2 & 7/2 & 9/2 & 11/2 \\
    1 & 1 & 1 & 1 & 1 & 1 
\end{bmatrix}
\begin{bmatrix}
    a_{1} \\
    a_{2} \\
    a_{3} \\
    a_{4} \\
    a_{5} \\
    a_{6}     
\end{bmatrix}
=
\begin{bmatrix}
    \sqrt{2 R_{LE}} \\
    z_u \\
    0 \\
    z_{xx_u} \\
    tan\theta_{TE_u} \\
    z_{TE_u}    
\end{bmatrix}
$$

The same relations apply to the lower surface, and the same matrix form can be used, except that the subscripts are changed from $u$ to $l$ to indicate the parameters relating to the lower surfaces. The $\sqrt{2 R_{LE}}$ is also negated for the lower surface.

```julia
import AirfoilParams
using Plots
pyplot()

#get x,y points for default NACA 2412 for comparision
xn,ynu,ynl = AirfoilParams.naca4()

#calculate the PARSEC parameters based on the NACA geometry
p = AirfoilParams.xytoparsec(xn,ynu,ynl)

#obtain the x,y points from the calculated PARSEC parameters
xp,ypu,ypl = AirfoilParams.parsec(p)

#plot a comparision
Plots.plot(xlim=(-0.1, 1.1), ylim=(-.25, .25),aspect_ratio=:equal)
Plots.plot!(xn,ynu,linecolor="black",label="NACA 2412")
Plots.plot!(xn,ynl,linecolor="black",label="")
Plots.plot!(xp,ypu,linecolor="red",linestyle=:dash,label="PARSEC")
Plots.plot!(xp,ypl,linecolor="red",linestyle=:dash,label="")
```
